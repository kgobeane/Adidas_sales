```python
sns.set_theme(style="whitegrid")
plt.rcParams['font.sans-serif'] = 'Arial'
plt.rcParams['font.family'] = 'sans-serif'
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df['Year_Month'] = df['Order_Date'].dt.to_period('M')

# Set a uniform visual theme
sns.set_theme(style="whitegrid")

# Create a 4x2 grid framework for all the subplots
fig, axes = plt.subplots(4, 2, figsize=(16, 22))

# --- Plot 1: Category Performance ---
category_data = df.groupby('Category')[['Revenue']].sum().sort_values(by='Revenue', ascending=False).reset_index()
sns.barplot(data=category_data, x='Category', y='Revenue', palette='viridis', ax=axes[0, 0])
axes[0, 0].set_title('Total Revenue by Product Category', fontsize=14, pad=10)
axes[0, 0].set_ylabel('Total Revenue ($)', fontsize=12)
axes[0, 0].set_xlabel('Category', fontsize=12)

# --- Plot 2: Regional Performance ---
region_data = df.groupby('Region')[['Revenue']].sum().sort_values(by='Revenue', ascending=False).reset_index()
sns.barplot(data=region_data, x='Revenue', y='Region', palette='magma', ax=axes[0, 1])
axes[0, 1].set_title('Total Revenue by Region', fontsize=14, pad=10)
axes[0, 1].set_xlabel('Total Revenue ($)', fontsize=12)
axes[0, 1].set_ylabel('Region', fontsize=12)

# --- Plot 3: Store Type Performance ---
store_data = df.groupby('Store_Type')[['Revenue']].sum().sort_values(by='Revenue', ascending=False).reset_index()
sns.barplot(data=store_data, x='Store_Type', y='Revenue', palette='coolwarm', ax=axes[1, 0])
axes[1, 0].set_title('Total Revenue by Store Type', fontsize=14, pad=10)
axes[1, 0].set_ylabel('Total Revenue ($)', fontsize=12)
axes[1, 0].set_xlabel('Store Type', fontsize=12)

# --- Plot 4: Customer Age Distribution ---
sns.histplot(data=df, x='Customer_Age', kde=True, bins=20, color='purple', ax=axes[1, 1])
axes[1, 1].set_title('Customer Age Distribution', fontsize=14, pad=10)
axes[1, 1].set_xlabel('Age', fontsize=12)
axes[1, 1].set_ylabel('Count', fontsize=12)

# --- Plot 5: Monthly Trends ---
monthly_data = df.groupby('Year_Month')[['Revenue', 'Profit']].sum().reset_index()
monthly_data['Year_Month'] = monthly_data['Year_Month'].astype(str)
monthly_data = monthly_data.sort_values(by='Year_Month')
axes[2, 0].plot(monthly_data['Year_Month'], monthly_data['Revenue'], marker='o', label='Revenue', color='b')
axes[2, 0].plot(monthly_data['Year_Month'], monthly_data['Profit'], marker='s', label='Profit', color='g')
axes[2, 0].set_title('Monthly Revenue & Profit Trend', fontsize=14, pad=10)
axes[2, 0].set_xlabel('Month', fontsize=12)
axes[2, 0].set_ylabel('Amount ($)', fontsize=12)
axes[2, 0].tick_params(axis='x', rotation=45)
axes[2, 0].legend()

# --- Plot 6: Payment Method Performance ---
payment_data = df.groupby('Payment_Method')[['Revenue']].sum().sort_values(by='Revenue', ascending=False).reset_index()
sns.barplot(data=payment_data, x='Revenue', y='Payment_Method', palette='crest', ax=axes[2, 1])
axes[2, 1].set_title('Total Revenue by Payment Method', fontsize=14, pad=10)
axes[2, 1].set_xlabel('Total Revenue ($)', fontsize=12)
axes[2, 1].set_ylabel('Payment Method', fontsize=12)

# --- Plot 7: Metric Correlation Heatmap ---
numerical_cols = ['Units_Sold', 'Unit_Price', 'Discount', 'Revenue', 'Profit', 'Customer_Age']
corr = df[numerical_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", square=True, ax=axes[3, 0])
axes[3, 0].set_title('Correlation Heatmap', fontsize=14, pad=10)

# --- Clean up the Grid Layout ---
# Turn off the empty 8th subplot frame space (row 4, column 2)
axes[3, 1].axis('off')

# Set dashboard wide super-title and configuration adjustments
fig.suptitle('Adidas Sales Analytics - EDA Dashboard', fontsize=20, weight='bold', y=0.99)
plt.tight_layout()

# Save the combined dashboard image frame
plt.savefig('adidas_eda_dashboard.png', bbox_inches='tight', dpi=150)
plt.close()

print("The unified dashboard frame 'adidas_eda_dashboard.png' has been generated and saved!")
```
